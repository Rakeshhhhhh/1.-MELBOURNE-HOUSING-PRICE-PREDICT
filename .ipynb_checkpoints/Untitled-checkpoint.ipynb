```python
def convert(x):
    if 'Sq. Meter' in x:
        x = x.replace('Sq. Meter','')
        x  = float(x) * 10.7639
    
    elif 'Perch' in x:
        x = x.replace('Perch','')
        x  = float(x) * 272.25
        
    elif 'Sq. Yards' in x:
        x = x.replace('Sq. Yards','')
        x  = float(x) * 9
        
    elif 'Acres' in x:
        x = x.replace('Acres','')
        x  = float(x) * 43560
        
    elif 'Cents' in x:
        x = x.replace('Cents','')
        x  = float(x) * 435.56
        
    elif 'Guntha' in x:
        x = x.replace('Guntha','')
        x  = float(x) * 1089
        
    elif 'Grounds' in x:
        x = x.replace('Grounds','')
        x  = float(x) * 2400
        
    elif '-' in x:
        x = x.split('-')
        if len(x)==2:
            x = (float(x[0]) + float(x[1]))/2
        else:
            x = float(x)  # for normal values eg- 56.98,7.9
        
    return x
```

### 5. Handle Outliers

**A)** As a data scientist when you have a conversation with your business manager (who has expertise in real estate), he will tell you that normally square ft per bedroom is 300 (i.e. 2 bhk apartment is minimum 600 sqft. If you have for example 400 sqft apartment with 2 bhk than that seems suspicious and can be removed as an outlier. We will remove such outliers by keeping our minimum thresold per bhk to be 300 sqft

```python
df4[(df4['total_sqft'] / df4['BHK']) < 300]
```

```python
df4 = df4[~((df4['total_sqft'] / df4['BHK']) < 300)]
df4.shape
```

**B)** Here we find that min price per sqft is 267 rs/sqft whereas max is 12000000, in Bangalore such a small price is just like a joke so it is wrong & such a high price may exist if the place is in prime/famous area but as we are making a generic model they will be like outliers for rest of the data and affect model's performance so remove them,

this shows a wide variation in property prices. **We should remove outliers per location using mean and one standard deviation**

```python
df4['location'].nunique()
```

```python
def remove(df,column):
    l = df[column].unique()
    df_out = pd.DataFrame(columns=df.columns)
    for i in l:
        df_n = df[df[column]==i]
        mean = df_n['price_per_sqft'].mean()
        std = df_n['price_per_sqft'].std()
        lower = mean-std
        higher = mean+std
        df_n = df_n[(df_n['price_per_sqft']>lower) & (df_n['price_per_sqft']<=higher)]
        df_out = pd.concat([df_out,df_n])
        
    return df_out

# def remove_pps_outliers(df):
#     df_out = pd.DataFrame()
#     for key, subdf in df.groupby('location'):
#         m = np.mean(subdf.price_per_sqft)
#         st = np.std(subdf.price_per_sqft)
#         reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
#         df_out = pd.concat([df_out,reduced_df],ignore_index=True)
#     return df_out
# df5 = remove_pps_outliers(df4)
# df5
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for BHK, bhk_df in location_df.groupby('BHK'):
            bhk_stats[BHK] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for BHK, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(BHK-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df6 = remove_bhk_outliers(df5)
# df8 = df7.copy()
df6.shape
```

# Feature scaling

as it's a regression problem so we need to do feature scaling otherwise features having high values will dominate

- **StandardScaler:** This scaler scales the features to have a mean of 0 and a standard deviation of 1. It is suitable when the distribution of your numeric features is approximately Gaussian (or normal). However, StandardScaler is sensitive to outliers, so if your numeric columns contain outliers, it might not be the best choice.

- **MinMaxScaler:** This scaler scales the features to a specified range, usually between 0 and 1. It preserves the shape of the original distribution and is less affected by outliers compared to StandardScaler. MinMaxScaler is a good choice when you have no prior knowledge of the distribution of your features or when you want to preserve the relationships between features.


- Given that your numeric columns have large values, it might be a good idea to use MinMaxScaler to scale them to a range between 0 and 1. This would prevent the large values from dominating the modeling process, especially if you are using algorithms sensitive to feature scales, such as K-Nearest Neighbors or Support Vector Machines.

- StandardScaler is sensitive to outliers, so if your numeric columns contain outliers, it might not be the best choice. MinMaxScaler is a good choice when you have no prior knowledge of the distribution of your features or when you want to preserve the relationships between features.

- However, if you are concerned about outliers in your numeric columns, you might consider robust scalers like RobustScaler, which are less influenced by outliers compared to StandardScaler.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

col=['total_sqft','bath','balcony','BHK']
train_x[col]=scaler.fit_transform(train_x[col])
train_x.head
```

```python
test_x[col]=scaler.transform(test_x[col])
test_x.head
```
